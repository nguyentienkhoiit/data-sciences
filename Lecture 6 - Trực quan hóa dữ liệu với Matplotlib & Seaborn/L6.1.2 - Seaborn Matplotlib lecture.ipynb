Matplotlib is a package for data visualization in Python. The library is built on NumPy arrays, and designed to work with the broader SciPy stack.

One of Matplotlib’s most important features is its **ability to play well with many operating systems and graphics backends**. Matplotlib supports dozens of backends and output types, which means you can count on it to work regardless of which operating system you are using or which output format you wish. This cross-platform, everything-to-everyone approach has been one of the great strengths of Matplotlib. It has led to a large user base, which in turn has led to an active developer base and Matplotlib’s powerful tools and ubiquity within the scientific Python world.

Matplotlib is  designed to help users to visualize data as easily as possible, with all the necessary control -- that is, by using relatively high-level commands most of the time, and still have the ability to use the low-level commands when needed.

Therefore, everything in matplotlib is organized in a hierarchy. At the top of the hierarchy is the matplotlib "state-machine environment" which is provided by the [`matplotlib.pyplot`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot) module. At this level, simple functions are used to add plot elements (lines, images, text, etc.) to the current axes in the current figure.

### Question 4
Among the week, which day has more mean of **tip**? 
- Visualize with a **barplot**.

✳️ There are 2 option to show bar labels:
+ option 1: using bar_label function.
+ option 2: using for loop

````python
dfplot = df.groupby('day')['tip'].mean().reset_index()
dfplot
````

````python
plt.figure(figsize = (7,4))
ax = sns.barplot(
    data = dfplot,
    x = 'day', 
    y = 'tip',
    order = ['Thur', 'Fri', 'Sat', 'Sun']
)
# option 1: using bar label:
plt.bar_label(ax.containers[0], fmt='%.2f')
plt.ylim(0, 4)
plt.title('MEAN OF TIP BY DAYS', loc = 'left')
plt.ylabel('Mean of tip')
plt.show()
````

````python
plt.figure(figsize = (7,4))
ax = sns.barplot(
    data = dfplot,
    x = 'day', 
    y = 'tip',
    order = ['Thur', 'Fri', 'Sat', 'Sun']
)
# option 2: using for loop
dfplot = dfplot.set_index('day').loc[['Thur', 'Fri', 'Sat', 'Sun']].reset_index()
for i, t in enumerate(dfplot.tip):
    plt.text(
        x = i,
        y = t + 0.05,
        s = str(round(t, 2)),
        ha = 'center'
    )
plt.ylim(0, 4)
plt.title('MEAN OF TIP BY DAYS', loc = 'left')
plt.ylabel('Mean of tip')
plt.show()
````

### Question 6

Male or Female who average `tip` more through days with 2 versions: `bar chart` and `line chart` 

````python
# your code here:
dfplot3 = df.groupby(['day', 'sex'])['tip'].mean().reset_index()
dfplot3['tip'] = round(dfplot3.tip, 2)
dfplot3
````

````python
# Using bar chart:
ax = sns.barplot(
    data = dfplot3,
    x = 'day',
    y = 'tip',
    hue = 'sex',
    order = ['Thur', 'Fri', 'Sat', 'Sun']
)
for i in range(dfplot3.sex.nunique()):
    plt.bar_label(ax.containers[i])
plt.ylim(0, 4.5)
plt.title('MEAN OF TIP THROUGH DAYS AND SEX')
plt.ylabel('Mean of tip')
plt.show()
````

````python
# Using line chart:
dfplot3 = dfplot3.set_index('day').loc[['Thur', 'Fri', 'Sat', 'Sun']].reset_index()

ax = sns.lineplot(
    data = dfplot3,
    x = 'day',
    y = 'tip',
    hue = 'sex'
)
for i, day_name in enumerate(dfplot3.day.unique()):
    female_mean_tip = dfplot3.loc[(dfplot3.day == day_name) & (dfplot3.sex == 'Female'), 'tip'].values[0]
    male_mean_tip = dfplot3.loc[(dfplot3.day == day_name) & (dfplot3.sex == 'Male'), 'tip'].values[0]
    plt.text(x = i, y = male_mean_tip + 0.02, s = str(male_mean_tip), ha = 'center')
    plt.text(x = i, y = female_mean_tip, s = str(female_mean_tip), ha = 'center')


plt.ylim(2, 4)
plt.title('MEAN OF TIP THROUGH DAYS AND SEX')
plt.ylabel('Mean of tip')
plt.show()
````

# **C. PANDAS BUILT-IN PLOT FUNCTION `.plot()`**

**When is it useful?**

☀︎ Quick visualization. I mean, QUICK!

☀︎ Line plot with time series data.

☀︎ Stacked bar, stacked area, stacked histogram.

**How to do it?** ▸ Simply get the plot table and call ```.plot()``` with ```kind = ``` to specify the kind of chart.

*DataFrame*`.plot(*args, **kwargs)`
```
Parameters:
data:   Series or DataFrame
        The object for which the method is called.

x:      label or position, default None
        Only used if data is a DataFrame.

y:      label, position or list of label, positions, default None
        Allows plotting of one column versus another. Only used if data is a DataFrame.
kind:   str, The kind of plot to produce:
            ‘line’ : line plot (default)
            ‘bar’ : vertical bar plot
            ‘barh’ : horizontal bar plot
            ‘hist’ : histogram
            ‘box’ : boxplot
            ‘kde’ : Kernel Density Estimation plot
            ‘density’ : same as ‘kde’
            ‘area’ : area plot
            ‘pie’ : pie plot
            ‘scatter’ : scatter plot (DataFrame only)
            ‘hexbin’ : hexbin plot (DataFrame only)
figsize: a tuple (width, height) in inches, Size of a figure object.
title:  Chart's title
```

When creating the ```plot_data``` table for plotting:

❗️ **The axis dimension must be index of the plot_data table**.
