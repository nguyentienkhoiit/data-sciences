```python
# YOUR CODE HERE
plt.figure(figsize=(12, 4))

dfplot = df.groupby('Month')['Profit'].sum().reset_index()
month_list_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.barplot(
    data = dfplot,
    x = 'Month',
    y = 'Profit', 
    order= month_list_order)

plt.title('Profit by Month')
plt.show()
```

**Question** This data was collected between 2011-01-01 and 2016-07-31. This means that the plot above was biased towards the months of January-July. So now your boss wants you to redraw the plot using only data of the years 2011-2015.

How is the new plot different from the old plot?

<ol type='A'>
    <li>The new plot shows that the profit tends to increase as it gets later in the year.</li>
    <li>The new plot no longer shows a significant spike of profit between November and December.</li>
    <li>The month with the least amount of profit is the same in both plots.</li>
    <li>There is no major differences between the two plots.</li>
</ol>

```python
# Plot the chart here
plt.figure(figsize=(12, 4))

df2011_2015 = df[df['Year'] <= 2015]
dfplot = df2011_2015.groupby('Month')['Profit'].sum().reset_index()
month_list_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.barplot(
    data = dfplot,
    x = 'Month',
    y = 'Profit', 
    order= month_list_order)

plt.title('Profit by Month from 2011-2015')
plt.show()
```

### Question 5 (10 pts)

```python
# Visualize the Ratio of Profit/Cost by Year.

# Expected Result
```

```python
# Step 1: T√≠nh t·ªïng profit v√† cost theo t·ª´ng nƒÉm
dfplot = df.groupby('Year')['Profit', 'Cost'].sum().reset_index()
dfplot
```

```python
# Step 2: T√≠nh t·ªâ l·ªá Profit/Cost:
dfplot['ppc_ratio'] = dfplot['Profit'] / dfplot['Cost']
dfplot
```

```python
# Step 2: Viz k·∫øt qu·∫£ c·ªßa dfplot
plt.figure(figsize=(7, 4))
sns.lineplot(
    data = dfplot,
    x = 'Year',
    y = 'ppc_ratio')

plt.title('Profit over Cost Ratio')
plt.ylabel('')
plt.show()
```

**Question:** Select the sub-categories of **Accessories** that have higher-than-average *profit per products sold*.

<ol type='A'>

```python
# your answer here
# B∆∞·ªõc 1: t√≠nh gi√° tr·ªã l·ª£i nhu·∫≠n (profit per products sold) tr√™n t·ª´ng sub_category:
df_acc2 = df_accessories.groupby('Sub_Category').agg({'Product': 'count', 'Profit': 'sum'}).reset_index()
df_acc2['profit_per_sold'] = df_acc2['Profit']/df_acc2['Product']
df_acc2
```

```python
# Plot revenue by AgeGroup

# In the 1st subplot: plot the Total Revenue by Age Group.
# In the 2nd subplot: plot the Average Revenue by Age Group.

# Expected Result
```

```python
# YOUR CODE HERE
plt.figure(figsize=(15, 4))
age_group_order = ['Youth (<25)', 'Young Adults (25-34)', 'Adults (35-64)', 'Seniors (64+)']
# Total revenue by age_group
plt.subplot(1, 2, 1)

dfplot1 = df.groupby('Age_Group')['Revenue'].sum().reset_index()
sns.barplot(
    data = dfplot1,
    x = 'Age_Group',
    y = 'Revenue',
    order = age_group_order
)
plt.ylabel('sum')
plt.title('Total Revenue')   # T√™n bi·ªÉu ƒë·ªì 1

# AVG revenue by age_group
plt.subplot(1, 2, 2)
dfplot2 = df.groupby('Age_Group')['Revenue'].mean().reset_index()
sns.barplot(
    data = dfplot2,
    x = 'Age_Group',
    y = 'Revenue',
    order = age_group_order
)
plt.ylabel('mean')
plt.title('Average Revenue')   # T√™n bi·ªÉu ƒë·ªì 1

plt.suptitle('REVENUE BY AGE GROUP')
plt.show()
```
