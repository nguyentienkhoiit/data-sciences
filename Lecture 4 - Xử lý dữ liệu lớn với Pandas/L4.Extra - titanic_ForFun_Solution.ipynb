 We can use column `Name` to predict passenger's sex.

```python
# extract PREFIX
name = 'Johnston, Miss. Catherine Helen "Carrie"'
name.split(', ')[1].split('.')[0]
```

```python
# 1 -- Get prefix column
def get_prefix(name):
    return name.split(', ')[-1].split('.')[0]
```

```python
# Apply to get prefix
titanic['NamePrefix'] = titanic['Name'].apply(get_prefix)
```

```python
# Get all prefix:
titanic['NamePrefix'].unique()
```

```python
# 2 -- Get gender from prefix
female = ['Mrs', 'Miss', 'Mme', 'Ms', 'Lady', 'Mlle', 'the Countess']

def check_prefix(prefix):
    if prefix in female:
        return 'female'
    else:
        return 'male'

titanic['NewSex'] = titanic['NamePrefix'].apply(check_prefix)
```

```python
titanic.head()
```

```python
# After you finished the previous step, one of your co-worker has found data of the gender
# You are curious about how good your prediction is compare to the true label
# Write a function to print out the accuracy score 
# Your code here:
def check_acc(row):
    if row['Sex'] == row['NewSex']:
      return 1
    else:
      return 0

titanic['check'] = titanic[['Sex', 'NewSex']].apply(check_acc, axis=1)
```
